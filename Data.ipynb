```python
import numpy as np
import pandas as pd
```

```python
col_names = ['index', 'image'] + list(range(1,11)) + ['semantic1', 'semantic2', 'challenge']
AVA = pd.read_csv('AVA.txt', sep=' ', names=col_names)
AVA['mean'] = AVA.apply(lambda row: sum(row[range(1,11)] * range(1,11))/sum(row[range(1,11)]), axis = 1)
AVA['score'] = AVA.apply(lambda row: row['mean']/10, axis = 1)
AVA['rounded_mean'] = AVA.apply(lambda row: int(np.round(row['mean']))-1, axis = 1)
AVA
```

```python
AVA.to_csv('AVA_full.csv')
```
